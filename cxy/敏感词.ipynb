```python
import pandas as pd
data = pd.read_csv('./data/minganci.csv')
data
```

```python
word_list = '药性、药效、抗药性、方子、处方、坐诊、挂号、配方、临床、临床经验、疗程、疫苗、抗体、服用、药量、剂量、疗程、解药、抗药性、抗体、复发、浓度、倍浓度、不需要配合运动或饮食、排毒、毒素、权威、功效、疗效、服用、诊断、疑难杂症、配制、研究、调配、软化、溶解、剥离、打散、转化、净化脂肪、将脂肪分离打散成游离状、死性脂肪，活性脂肪、档案室、工作室、档案室、研究室、脂肪膨胀'
word_list = word_list.split('、')
word_list = list(set(word_list))
```

```python
import json
for i in range(len(data)):
    talk = json.loads(data.loc[i,'transData.sentenceList'].replace("'",'"'))
    print(i,data.loc[i,'correctInfoData.correctResult'])
    for chat in talk:
        for word in word_list:
            if word in chat['content'] and chat['role']=='AGENT':
                text = chat['content'].replace(word,"【"+word+"】")
                print(chat['role'],text)
    print()
```
