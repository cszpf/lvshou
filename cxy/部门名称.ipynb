```python
import pandas as pd
data = pd.read_csv('./data/bumenmingcheng.csv')
data
```

```python
word_list = '效果调查中心、效果调查科、专案组、疑难瘦身组、问题处理组、效果中心、疑难杂症研究中心、公司疑难瘦身专案组、数据调查部、调查小组、结案、绿瘦效果回访中心、资深组、资深顾问组、资深瘦身组资深顾问团、瘦身组、绿瘦减肥中心、助理团、资深效果组、助理办总部效果组、总部瘦身组、资深组办、资深小组、资深人士组、资深顾问办、资深办、专家组、专家顾问组、专案组、助理处 效果组、效果管理总部、数据组、瘦身专业组、瘦身专案组、瘦身团、审核组、审核秘书办、绿瘦效果组、绿瘦高级资深处、健康组、回访组、减肥人士组'
word_list = word_list.split('、')
word_list = list(set(word_list))
```

```python
import json
for i in range(len(data)):
    talk = json.loads(data.loc[i,'transData.sentenceList'].replace("'",'"'))
    print(i,data.loc[i,'correctInfoData.correctResult'])
    for chat in talk:
        for word in word_list:
            if word in chat['content'] and chat['role']=='AGENT':
                text = chat['content'].replace(word,"【"+word+"】")
                print(chat['role'],text)
    print()
```
